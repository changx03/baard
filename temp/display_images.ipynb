```python
import os
from pathlib import Path

import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision as tv
import torchvision.transforms.functional as tvF
```

```python
PATH_ROOT = Path(os.getcwd()).absolute().parent
print(PATH_ROOT)
```

```python
def show_top5_imgs(path_dataset,figsize=(8, 3), cmap='gray'):
    dataset = torch.load(path_dataset)
    images = [x for x, y in dataset]
    labels = [y for x, y in dataset]
    grid = tv.utils.make_grid(images[:5], nrow=5, padding=4, pad_value=1)
    grid = grid.permute(1,2,0)
    
    plt.figure(figsize=figsize)
    plt.imshow(grid.numpy(), cmap=cmap)
    plt.axis('off')
    plt.show()
    
    lbl_str = ', '.join([str(l.item()) for l in labels[:5]])
    print(f'Labels: {lbl_str}')
```

```python
def plot_images(data_name, lnorm, eps_list, att='APGD'):
    path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', data_name, 'AdvClean.n_100.pt')
    show_top5_imgs(path_img, cmap=None)
    print('Clean images')

    for eps in eps_list:
        path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', data_name, f'{att}.{lnorm}.n_100.e_{eps}.pt')
        show_top5_imgs(path_img, cmap=None)
        print(f'{att} {lnorm} eps={eps}')
```

```python
data_name = 'MNIST'
eps_list = [0.22, 0.28, 0.34, 0.41, 0.47, 0.53, 0.66, 0.72, 0.78, 0.84, 0.91, 0.97]
plot_images(data_name, 'Linf', eps_list)
```

```python
eps_list = [4, 6, 8, 10, 12, 14, 16]
plot_images(data_name, 'L2', eps_list)
```
