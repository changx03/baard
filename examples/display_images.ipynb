```python
import os
from pathlib import Path

import torch
import numpy as np
import matplotlib.pyplot as plt
import torchvision as tv
import torchvision.transforms.functional as tvF
```

```python
PATH_ROOT = Path(os.getcwd()).absolute().parent
print(PATH_ROOT)
```

```python
def show_top5_imgs(path_dataset,figsize=(8, 3), cmap='gray'):
    dataset = torch.load(path_dataset)
    images = [x for x, y in dataset]
    labels = [y for x, y in dataset]
    grid = tv.utils.make_grid(images[:5], nrow=5, padding=4, pad_value=1)
    grid = grid.permute(1,2,0)
    
    plt.figure(figsize=figsize)
    plt.imshow(grid.numpy(), cmap=cmap)
    plt.axis('off')
    plt.show()
    
    lbl_str = ', '.join([str(l.item()) for l in labels[:5]])
    print(f'Labels: {lbl_str}')
```

```python
path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', 'MNIST', 'APGD.L2.n_100.e_1.pt')
show_top5_imgs(path_img, cmap=None)

path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', 'MNIST', 'APGD.L2.n_100.e_16.pt')
show_top5_imgs(path_img, cmap=None)

path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', 'MNIST', 'CW2.L2.n_100.e_0.pt')
show_top5_imgs(path_img, cmap=None)

path_img = os.path.join(PATH_ROOT, 'results', 'exp1234', 'MNIST', 'CW2.L2.n_100.e_10.pt')
show_top5_imgs(path_img, cmap=None)
```
